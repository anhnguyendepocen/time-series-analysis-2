**Exercise 3.1**.  Verify Equation (3.3.2) on page 30, for the least square estimates of $\beta_0$ and $\beta_1$ when the model $Y_t = \beta_0 + \beta_1 t + X_t$ is considered.

**Solution**.

$$ 
\hat{\beta}_1 = \frac{\sum_{t=1}^n(Y_t - \overline{Y})(t - \overline{t})}{\sum_{t=1}^n (t - \overline{t})^2}
\quad
\hat{\beta}_0 = \overline{Y} - \hat{\beta}_1 \overline{t}
\tag{3.3.2}
$$

The chosen values should minimize

$$ Q(\beta_0, \beta_1) = \sum_{t=1}^n [Y_t - (\beta_0 + \beta_1 t)]^2 $$

Computing the partial derivative with respect to $\beta_0$,

$$ \frac{d Q}{d \beta_0} = \sum_{t=1}^n [-2 (Y_t - \beta_1 t) + 2 \beta_0] = -2 n (\overline{Y} + \beta_1 \overline{t}) + 2 n \beta_0 $$

so the quadratic error is minimized when this expression is 0, or

$$ \hat{\beta}_0 = \overline{Y} - \hat{\beta}_1 \overline{t} $$

Computing the partial derivative with respect to $\beta_1$,

$$ \frac{d Q}{d \beta_1} = \sum_{t=1}^n -2(Y_t t - \beta_0 t - \beta_1 t^2) $$
m
Substituting the estimate for $\hat{\beta}_0$ and equating the derivative to 0, we get

$$ \sum_{t=1}^n (Y_t t - \overline{Y}t + \hat{\beta}_1 \overline{t} t - \hat{\beta_1} t^2) = 0 $$

Separating into two sums,

$$ \sum_{t=1}^n t (Y_t - \overline{Y}) - \hat{\beta}_1 \sum_{t=1}^n (t^2 - \overline{t} t) = 0 $$

where, when isolating $\hat{\beta}_1$, gives the result

$$ \hat{\beta}_1 
= \frac{\sum_{t=1}^n (t Y_t - t\overline{Y})}{\sum_{t=1}^n (t^2 - t\overline{t})} 
= \frac{\sum_{t=1}^n (t Y_t - t\overline{Y}) + \sum_{t=1}^n (\overline{t} \overline{Y} - Y_t \overline{t})}{\sum_{t=1}^n (t^2 - t\overline{t}) + \sum_{t=1}^n (\overline{t}^2 - t \overline{t})} 
= \frac{\frac{1}{n}\sum_{t=1}^n(Y_t - \overline{Y})(t - \overline{t})}{\frac{1}{n}\sum_{t=1}^n (t - \overline{t})^2} 
= \frac{\sum_{t=1}^n(Y_t - \overline{Y})(t - \overline{t})}{\sum_{t=1}^n (t - \overline{t})^2}
$$

The algebraic manipulation becomes somewhat more intuitive using matrix notation,

$$ S(\beta) = (y - X \beta)^T (y - X \beta) $$

is minimized when 

$$ 0 = \frac{dS}{d\beta}(\hat{\beta}) = \frac{d}{d\beta} \left( y^T y - \beta^T X^T y - y^T X \beta + \beta^T X^T X \beta \right) |_{\beta=\hat{\beta}} = -2 X^T y + 2 X^T X \hat{\beta} $$

Assuming matrix $X$ has full column rank, then $X^T X$ is invertible, and so

$$ \hat{\beta} = (X^T X)^{-1} X^T y $$

**Exercise 3.2**.  Suppose $Y_t = \mu + e_t - e_{t - 1}$.  Find $\text{Var}[\overline{Y}]$.  Note any unusual results.  In particular, compare your answer to what would have been obtained if $Y_t = \mu + e_t$.  (Hint: You may avoid Equation (3.2.3) on page 28 by first doing some algebraic simplication on $\sum_{t=1}^n (e_t - e_{t - 1})$.)

**Solution**.

$$ \text{Var}[\overline{Y}] = \text{Var}\left[ \frac{1}{n} \sum_{t=1}^n Y_t \right] = \text{Var}\left[ \frac{1}{n} \sum_{t=1}^n (\mu + e_t - e_{t-1}) \right] = \text{Var}\left[\mu + \frac{1}{n}(e_{n} - e_{0})\right] = \text{Var}\left[\frac{e_{n}}{n}\right] = \frac{\sigma_e^2}{n^2}$$

If instead we had $Y_t = \mu + e_t$, the result would have been

$$ \text{Var}[\overline{Y}] = \text{Var}\left[ \frac{1}{n} \sum_{t=1}^n Y_t \right] = \text{Var}\left[ \frac{1}{n} \sum_{t=1}^n (\mu + e_t) \right] = \text{Var}\left[\frac{1}{n} \sum_{t=1}^n (e_t) \right] = \frac{1}{n^2} \sum_{t=1}^n \text{Var}[e_t] = \frac{\sigma_e^2}{n}$$

The autocovariance in the error terms meant that the variance of the series average now decays in $O(n^{-2})$ rather than $O(n^{-1})$ with white noise error.

**Exercise 3.3**.  Suppose $Y_t = \mu + e_t + e_{t-1}$.  Find $\text{Var}[\overline{Y}]$.  Compare your answer to what would have been obtained if $Y_t = \mu + e_t$.  Describe the effect that the autocorrelation in $\{Y_t\}$ has on $\text{Var}[\overline{Y}]$.

**Solution**.

The variance of $Y_t$ is

$$ \gamma_0 = \text{Var}[Y_t] = \text{Var}[e_t + e_{t - 1}] = 2 \sigma_e^2 $$

and the autocovariance for $k > 0$ is

$$ \gamma_k = \text{Cov}[Y_t, Y_{t - k}] = \begin{cases} 
\text{Cov}[e_{t - 1}, e_{t - 1}] = \sigma_e^2 & \text{for } k = 1 \\
0 &\text{for } k > 1
\end{cases}
$$

so the autocorrelation is

$$ \rho_k = \begin{cases}
1 &\text{for } k = 0 \\
1/2 &\text{for } k = 1 \\
0 &\text{for } k > 1
\end{cases} $$

From equation 3.2.3,

$$ \text{Var}[\overline{Y}] = \frac{\gamma_0}{n} \left[ 1 + 2 \sum_{k=1}^{n-1} \left( 1 - \frac{k}{n}\right) \rho_k \right] $$

and in our case

$$ \text{Var}[\overline{Y}] = \frac{2 \sigma_e^2}{n} \left[ 1 + 2 \left(1 - \frac{1}{n}\right) \frac{1}{2} \right] = \sigma_e^2 \frac{2n - 1}{n^2} $$

This also decreases as $O(n^{-1})$, but for sufficiently large $n$ the sample variance is twice as large as it would have been for $Y_t = \mu + e_t$.

**Exercise 3.4**.  The data file `hours` contains monthly values of the average hours worked per week in the U.S. manufacturing sector for July 1982 through June 1987.

**(a)**  Display and interpret the time series plot for these data.

**(b)**  Now construct a time series plot that uses separate plotting symbols for the various months.  Does your interpretation change from part (a)?

**Solution**.

**(a)**

```r
require(ggplot2)
```

```r
hours <- read.table("../data/hours.dat", header=TRUE)[['hours']]
options(repr.plot.width=12, repr.plot.height=4)

xx = 1:length(hours)/12 + 1982 + 7/12
ggplot() + 
  geom_line(aes(x=xx, y=hours), color='blue') +
  geom_point(aes(x=xx, y=hours), color='blue') +
  xlab('Year') + ylab('Hours') +
  theme_bw() + theme(text = element_text(size=16))
```

The seasonal patterns are not readily apparent when plotting the data this way, though the smoothness across related data points suggest autocorrelation, and there seems to be an overall trend going up.

**(b)**

```r
labels = c('Jan', 'Feb', 'Mar', 'Apr',
  'May', 'Jun', 'Jul', 'Aug',
  'Sep', 'Oct', 'Nov', 'Dec')

palette = c('#e2d9e2', '#b3c6ce', '#7ba1c2', '#6276ba',
 '#5e43a5', '#4e186f', '#2f1436', '#581647',
 '#8e2c50', '#b25652', '#c68a6d', '#d4bcac')
```

```r
hours <- read.table("../data/hours.dat", header=TRUE)[['hours']]
options(repr.plot.width=12, repr.plot.height=4)

month_index = ((1:length(hours)) + 5) %% 12 + 1
xx = 1:length(hours)/12 + 1982 + 7/12
ggplot() + 
  geom_line(aes(x=xx, y=hours), color='blue', alpha=0.2) +
  geom_text(aes(x=xx, y=hours, label=labels[month_index]), color=palette[month_index], size=4) + 
  xlab('Year') + ylab('Hours') +
  theme_bw() + theme(text = element_text(size=16))
```

Seasonality is much more apparent now: rise towards December and decay in January / February, and decays in July.
